Pandas Data Frame:

Pandas DataFrame is two-dimensional tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
import numpy as np
```

Creaating a Pandas DataFrame

```python
# importing the boston house price data
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
type(housing)
```

```python
print(housing)
```

```python
# pandas DataFrame
df = pd.DataFrame(housing.data , columns = housing.feature_names)
```

```python
df.head()
```

```python
df.shape
```

Manipulating a DataFrame

```python
df['Target_value'] = housing.target
```

```python
df.head()
```

```python
#removing a row
df.drop(index =3 , axis =0)
```

```python
#drop a column
df.drop(columns='AveRooms', axis =1)
```

```python
#এটা DataFrame থেকে 3rd row (index position 2) return করে।
df.iloc[2]
```

```python
# locating a particular column
df.iloc[:,1]  # second column 
df.iloc[:,2]   # third colum
df.iloc[:,-1]   #last column
```
