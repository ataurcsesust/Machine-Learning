Label Encoding:
- converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
iris = pd.read_csv('E:/ML/iris_data.csv')
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
label_encoder = LabelEncoder()
```

```python
iris_label =label_encoder.fit_transform(iris.Species)
```

```python
iris['target'] = iris_label
```
