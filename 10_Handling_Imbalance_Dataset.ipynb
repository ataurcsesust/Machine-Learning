**Imbalanced Dataset:**
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('E:/ML/credit_data.csv')
```

```python
df.head()
```

```python
df['Class'].value_counts()
```

This is Highly Imbalanced Dataset

```python
legid = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legid.shape)
print(fraud.shape)
```

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 492

```python
legid_sample = legid.sample(n=492)
```

```python
print(legid_sample.shape)
```

**Concatenate the Two DataFrames**

```python
new_df = pd.concat([legid_sample,fraud], axis =0)
```

```python
new_df.head()
```
