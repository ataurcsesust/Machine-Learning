```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

Plotting a Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units ='xy' ,angles ='xy' ,scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

সাধারণত quiver() vector-এর দৈর্ঘ্য ছোট/বড় করে দেখাতে পারে। কিন্তু scale=1 দিলে vector-এর প্রকৃত মান ব্যবহার করা হয়।

scale_units='xy'
Vector-এর দৈর্ঘ্য x-y axis-এর unit অনুযায়ী নির্ধারণ করবে।
অর্থাৎ (4,5) vector হলে সেটি সত্যিই x দিকে 4 unit এবং y দিকে 5 unit যাবে।
angles='xy'
Vector-এর direction x-y coordinate system অনুযায়ী গণনা করবে।
অর্থাৎ (4,5) মানে:
x দিকে 4
y দিকে 5

```python
plt.quiver(0,0,4,5, scale_units = 'xy' , angles= 'xy' , scale = 1 , color ='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-6, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two Vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])
sum = v1 + v2 
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units = 'xy', angles = 'xy' ,scale = 1 , color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy', angles = 'xy', scale =1 , color = 'y')
plt.quiver(0,0,5,1,scale_units = 'xy', angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of 2 Vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,3,-2])
diff = v1 - v2 
print(diff)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
v1 = np.asarray([0,0,2,3])
v2 = 2*v1
print(v2)
```

```python
plt.quiver(0,0,2,3,scale=1, color='g')
plt.quiver(0,0,4,6, scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
v1 = np.asarray([0,0,2,3])
v2 = -0.5*v1
print(v2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product of 2 Vectors

```python
a = np.array([2,3,4])
b = np.array([2,3,4])

a_dot_b = np.dot(a,b)

print(a_dot_b)
```

Cross Product of 2 Vectors

```python
c = np.array([5, 10, 20])

d = np.array([18, 32, 50])

c_cross_d =np.cross(c,d)

print(c_cross_d)
```

Projection of "a" vector on "v" Vector

```python
a = np.array([23, 45, 62])
v = np.array([45, 82, 67])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(np.sum(v**2))

#   (a*v)/(|v|^2)*v
projection_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector =', projection_of_a_on_v)

```
