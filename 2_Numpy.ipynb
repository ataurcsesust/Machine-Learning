Initial Placeholders in numpy arrays
NumPy-তে আমরা অনেক সময় শুরুতেই কিছু খালি/ডিফল্ট value দিয়ে filled array তৈরি করি। এগুলোকে initial placeholders বলে।

```python
# create a numpy array of Zeros
x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of ones
y = np.ones((3,3))
print(y)
```

```python
# array of a particular value
z = np.full((4,5),10)
print(z)
```

```python
# create an identity matrix

a = np.eye(5)
print(a)
```

```python
# create a numpy array with random values
b = np.random.random((4,3))
print(b)

```

```python
c = np.random.randint(10,20,(4,3))  #20 exclusive
print(c)
```

```python
# array of evenly spaced values --> specifying the number of values required , end value included
d = np.linspace(10,30,5)
print(d)
```

```python
# array of evenly spaced values --> specifying the step , end value exlueded
e = np.arange(10,30,5)
print(e)
```

```python
# convert a list to a numpy array
list2 = [10,20,20,20,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

Mathematical operations on a np array

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
#transpose
trans = array.T
print(trans)
print(trans.shape)
```

```python
# reshaping a array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
c = np.array([1,2,3,4,5,6])
b = c.reshape(3,2)
print(b)
print(b.shape)
```
