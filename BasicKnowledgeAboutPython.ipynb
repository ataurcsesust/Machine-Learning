Mutable → তৈরি হওয়ার পর পরিবর্তন করা যায়।

Immutable Objects:

List

```python
# list should be included in the square brackets
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# lists can have multiple data types
my_list = [2, 3, 1.8, 'English', True]
print(my_list)
```

```python
# add elements to a list
my_list = [2, 3, 1.8, 'English', True]
my_list.append(6)
print(my_list)
```

```python
# print elements of a list using their index
print(my_list[0])
print(my_list[2])
print(my_list[-1])  #last element of the list
```

```python
# lists allow duplicate values
list_1 = [1,2,3,4,5,12,2,3]
print(list_1)

print(len(list_1))
```

```python
# initiating an empty list
list_2 = []
print(list_2)
```

```python
list_2.append(5)
print(list_2)
```

```python
# delete an item in a list
list_2 = [2, 3, 1.8, 'English', True, 6]
print(list_2)

del list_2[3]
print(list_2)
```

```python
# join two lists
list_3 = [1,2,3,4,5]
list_4 = [6,7,8,9,10]

list_5 = list_3 + list_4
print(list_5)
```

Tuple

```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
# tuple allows multiple data types
tuple_2 = (1,2,3.5, 'Machine Learning', False)
print(tuple_2)
```

```python
# converting  a list to a tuple
my_list = [1,2,3,4]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[1])
```

```python
#Tuples are immutable --> Unchangeable
my_tuple.append(6)
```

```python
print(len(my_tuple))
```
