Conversion of one data type to another

```python
# int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
# float to int
x = 5.88
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```
