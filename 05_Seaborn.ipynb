Seaborn:
- Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# setting a theme for the plots
# এটি Seaborn-এর default theme/style activate করে, যাতে graph গুলো আরও সুন্দর ও professional দেখায়।
sns.set_theme()
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill',y='tip', col='time',hue='smoker',style='smoker',size ='size')
#col='time'  আলাদা আলাদা plot তৈরি করবে:
#hue='smoker'  color দিয়ে differentiate করবে:
#style='smoker'  marker shape পরিবর্তন করবে:
#dataset-এর size নামের column অনুযায়ী point-এর size (marker size) পরিবর্তন হবে।
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house = pd.DataFrame(housing.data, columns=housing.feature_names)
house['Price'] = housing.target
```

```python
print(housing)
```

```python
house.head()
```

```python
sns.histplot(house['Price'])
```

Correlation:
- Positive Correlation
- Negative Correlation


```python
correlation = house.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation , cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')

#cbar=True পাশে color bar দেখাবে
#square=True সব cell square shape হবে
#fmt='.1f'  number 1 decimal place এ দেখাবে
#annot=True প্রতিটা box-এর ভিতরে value দেখাবে
#annot_kws={'size':8}           text size ছোট করবে
#cmap='Blues' blue color theme ব্যবহার করবে
```
