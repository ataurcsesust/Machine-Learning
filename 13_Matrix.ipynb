Creating Matrices with Random Values

```python
random_mat = np.random.rand(3,3)
#random_mat = np.random.random((3,3)) # It's do the same work just syntax different
print(random_mat)
```

```python
# creating matrix with random integers
random_integer_matrix = np.random.randint(100 , size=(4,5))
#random_integer_matrix = np.random.randint( 20,100 , size=(4,5)) do same work but integer range 20 to 100
print(random_integer_matrix)
```

Matrix with all the values as 1

```python
matrix_3 = np.ones((2,3))
print(matrix_3)
```

```python
matrix_3 = np.ones((2,3),dtype =int)
print(matrix_3)
```

Null Matrix or Zero Matrix

```python
null_matrix = np.zeros((4,4))
print(null_matrix)
```

```python
null_matrix = np.zeros((4,4), dtype = int)
print(null_matrix)
```

Identity Matrix

```python
identity_matrix = np.eye(4,4)
print(identity_matrix)
```

```python
identity_matrix = np.eye(3,3)
print(identity_matrix)
```

```python
identity_matrix = np.eye(5, dtype = int)
print(identity_matrix)
```

Transpose of a Matrix

```python
# matrix with random integer values
a = np.random.randint(100, size=(4,5))
print(a)
```

```python
tranpose = np.transpose(a)
#tranpose = a.T # this is also use for tranpose Matrix
print(tranpose)
```

Matrix Addition

```python
# creating two matrices

A = np.array([[2,3],[4,5]])

B = np.array([[6,7],[8,9]])

#sum = A + B  # It's do the same work
sum = np.add(A,B)
print(sum)
```

```python
# create two matrices with random values
matrix_3 = np.random.randint(10, size=(3,3))
matrix_4 = np.random.randint(20, size=(3,4))
sum_2 = np.add(matrix_3, matrix_4)
```

**Matrix Subtraction**

```python
# creating two matrices

A = np.array([[2,3],[4,5]])

B = np.array([[6,7],[8,9]])

#diff = A - B   # It's also do the same work
diff = np.subtract(A,B)

print(diff)
```

**Multiplying a matrix by a scalar**

```python
x = 5
y = np.random.randint(10, size=(4,4))

print(y)
```

```python
product = np.multiply(x,y)

print(product)
```

**Multiplying 2 Matrices**

```python
# create two matrices with random values
matrix_3 = np.random.randint(5, size=(3,3))
matrix_4 = np.random.randint(5, size=(3,3))
print(matrix_3)
print(matrix_4)
```

```python
product = np.dot(matrix_3, matrix_4)
print(product)
```

Element wise Multiplication

```python
product = np.multiply(matrix_4,matrix_3)
print(product)
```
